```python
# Program to calculate accuracy based on original test MSE corresponding to TL and predicted test MSE 
# corresponding to TL using CNN (similarity measure).
```

```python
import pandas as pd 
```

### Original Part

```python
# loading csv file having original test MSE corresponding to TL for wireless environments.

df_org = pd.read_csv('results_indoor_env_TL_all_main.csv')
```

```python
df_org.head()
```

```python
# Taking columns of original test MSE corresponding to TL only.

df1=df_org[['mse_test_0.95', 'mse_test_0.9', 'mse_test_0.85', 'mse_test_0.8', 
        'mse_test_0.75', 'mse_test_0.7', 'mse_test_0.65', 'mse_test_0.6']] 
```

```python
df1.head()
```

```python
df1.shape
```

```python
df1.describe()
```

```python
# Taking threshold for original test MSE corresponding to TL and checking each row.
# If test MSE value is less than threshold then decision-yes, else decision-no

dictn1 = {
    'decision_org_mse': []    
}

mse_th = 0.004  # threshold for original test MSE corresponding to TL

# loop for MSE
for idx in range(len(df1)):
    for j, i in enumerate(df1.columns):
        go = False
        if i.startswith('mse_test'):
            if df1.loc[idx, i] < mse_th:
                go = True
                dictn1['decision_org_mse'].append('Yes')
                break
                
        if go == False and j == len(df1.columns) - 1:
            dictn1['decision_org_mse'].append('No')
```

```python
res1 = pd.DataFrame(dictn1) # Results dataframe for original test MSE
# res1
```

```python
res1.head()
```

```python
# To see number of wireless environments performing/not performing TL

res1['decision_org_mse'].value_counts()
```

```python
# To see which wireless environments are agreeing with the threshold value.

# res1[res1['decision_org_mse'] == 'Yes']  # Yes/No
```

### Predicted Part

```python
# loading csv file having predicted test MSE corresponding to TL obtained from CNN similarity measure.

df_pred = pd.read_csv('results_y_real_inv.csv') # predicted test MSE corresponding to TL obtained from CNN.
```

```python
df_pred.head()
```

```python
df2=df_pred.iloc[:, 1:10] # Taking columns of predicted test MSE corresponding to TL only 
```

```python
df2.head()
```

```python
df2.describe()
```

```python
# Taking threshold for predicted test MSE corresponding to TL and checking each row
# If MSE value is less than threshold then decision-yes, else decision-no

dictn2 = {
    'decision_pred_mse': []    
}

mse_th = 0.00384  # threshold for predicted test MSE corresponding to TL (because 4 % error in CNN training)

# loop for MSE
for idx in range(len(df2)):
    for j, i in enumerate(df2.columns):
        go = False
        if i.startswith('mse_test'):
            if df2.loc[idx, i] < mse_th:
                go = True
                dictn2['decision_pred_mse'].append('Yes')
                break
                
        if go == False and j == len(df2.columns) - 1:
            dictn2['decision_pred_mse'].append('No')
```

```python
res2 = pd.DataFrame(dictn2) # Results dataframe for predicted test MSE
# res2
```

```python
res2.head()
```

```python
# To see number of wireless environments performing/not performing TL

res2['decision_pred_mse'].value_counts()
```

```python
# To see which wireless environments is agreeing with the threshold value.

# res2[res2['decision_org_mse'] == 'Yes']  # Yes/No
```

### Accuracy Calculation

```python
# concatenating decisions of both original and predicted dataframes

final_df = pd.concat([res1, res2], axis=1)
# final_df
```

```python
final_df.head()
```

```python
# Finding TP, TN, FP and FN

dictn = {
    'output': []
}

for i in range(len(final_df)):
    if final_df.loc[i, 'decision_org_mse'] == 'Yes' and final_df.loc[i, 'decision_pred_mse'] == 'Yes':
        dictn['output'].append('TP')
        
    elif final_df.loc[i, 'decision_org_mse'] == 'No' and final_df.loc[i, 'decision_pred_mse'] == 'No':
        dictn['output'].append('TN')
        
    elif final_df.loc[i, 'decision_org_mse'] == 'Yes' and final_df.loc[i, 'decision_pred_mse'] == 'No':
            dictn['output'].append('FN')
    
    elif final_df.loc[i, 'decision_org_mse'] == 'No' and final_df.loc[i, 'decision_pred_mse'] == 'Yes':
            dictn['output'].append('FP')
```

```python
result = pd.DataFrame(dictn)
# result
```

```python
result.head()
```

```python
# Counting number of TP, TN, FP and FN

m = {
    'TP': 0,
    'TN': 0,
    'FP': 0,
    'FN': 0
}

for i in range(len(result)):
    if result.loc[i, 'output'] == "FN":
        m['FN'] += 1
    elif result.loc[i, 'output'] == "FP":
        m['FP'] += 1
    elif result.loc[i, 'output'] == "TN":
        m['TN'] += 1
    elif result.loc[i, 'output'] == "TP":
        m['TP'] += 1
```

```python
m # counter # FP and FN should be as less as possible.
```

```python
prec = m['TP'] / (m['TP'] + m['FP'])
recall = m['TP'] / (m['TP'] + m['FN'])
f1_score = (2*prec*recall)/(prec + recall)
```

```python
print("Precision: ", prec)
print("Recall: ", recall)
print("F1-score: ", f1_score)
```

```python
acc = (m['TN'] + m['TP']) / (m['TN'] + m['TP'] + m['FN'] + m['FP'])
print("Accuracy: ", acc)
```

```python
sens = m['TP'] / (m['TP'] + m['FN'])
print("Sensitivity/Transferability: ", sens)
```
